**Gemini Deep Research** is an agentic workflow that autonomously plans, executes, and synthesizes multi-step research tasks. Powered by **Gemini 3.1 Pro**, it navigates complex information landscapes using web search and your own data to produce detailed, cited reports.

---

> **Note:** This notebook uses the `deep-research-pro-preview-12-2025` agent and requires a Gemini API key.

## ⚙️ Setup
Initialize the Gemini API client. Make sure you have obtained an API key from [Google AI Studio](https://aistudio.google.com/app/apikey) and added it to your Colab secrets.

```python
import time

from google import genai
from google.colab import userdata

# Securely retrieve API Key
# Make sure to add 'GEMINI_API_KEY' to your Colab Secrets (key icon on the left)
GEMINI_API_KEY = userdata.get("GEMINI_API_KEY")

client = genai.Client(api_key=GEMINI_API_KEY)
```

## 📝 Research Parameters
Define your research topic and optionally provide a [File Search Store](https://ai.google.dev/api/file-search/file-search-stores) to include your own documents in the research.

```python
# @title Research Input
RESEARCH_TOPIC = "Research the history and future of Solid State Batteries." # @param {type:"string"}
# @markdown Optional: Enter a File Search Store name (e.g., `projects/.../locations/.../fileSearchStores/...`) to use your own documents.
FILE_SEARCH_STORE_NAME = "" # @param {type:"string"}
```

## 🚀 Run Research
This cell starts the research agent, leveraging the [Gemini API](https://ai.google.dev/gemini-api) to autonomously browse the web and your files to compile a comprehensive report. This process can take several minutes.

```python
def poll_interaction(interaction_id):
    """Polls the interaction status until complete or failed."""
    while True:
        try:
            interaction = client.interactions.get(interaction_id)
            if interaction.status == "completed":
                return interaction
            if interaction.status == "failed":
                return interaction

            time.sleep(10)
        except Exception:
            time.sleep(10)


tools = []
if FILE_SEARCH_STORE_NAME:
    tools.append({
        "type": "file_search",
        "file_search_store_names": [FILE_SEARCH_STORE_NAME],
    })

# Start the research task in the background
interaction = client.interactions.create(
    input=RESEARCH_TOPIC,
    agent="deep-research-pro-preview-12-2025",
    background=True,
    tools=tools if tools else None,
)

interaction = poll_interaction(interaction.id)
```

## 📊 Research Report
Once the agent finishes, the report is displayed below using [IPython Markdown](https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html) format.

```python
from IPython.display import Markdown

if interaction.status == "completed":
    # Display the final report using Markdown for better formatting
    display(Markdown(interaction.outputs[-1].text))
else:
    pass
```

## 💬 Follow-up Questions
You can dive deeper into specific parts of the report by asking follow-up questions using the [Gemini standard models](https://ai.google.dev/gemini-api/docs/models/gemini).

```python
# @title Follow-up Question
FOLLOW_UP_QUESTION = "Can you elaborate on the key findings?" # @param {type:"string"}
```

```python
if "interaction" in locals() and interaction.status == "completed":

    # Follow-up interactions use the standard model but reference the previous agent interaction
    follow_up_interaction = client.interactions.create(
        input=FOLLOW_UP_QUESTION,
        model="gemini-3.1-pro-preview",
        previous_interaction_id=interaction.id,
    )

    display(Markdown(follow_up_interaction.outputs[-1].text))
else:
    pass
```
